```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
def actual_y(x):
    random_val = np.random.random(x.shape) * 1000 - 500
    return 2 * (x**2) + 3 * x -5 + random_val
```

```python
x = np.array([[i] for i in range(-50, 50, 1)])
y = actual_y(x)
y
```

```python
plt.scatter(x,y)
```

# Implementing through linear regresion

```python
from sklearn.linear_model import LinearRegression
```

```python
clf = LinearRegression()
clf.fit(x, y)
y_pred_1 = clf.predict(x)
```

```python
plt.scatter(x,y,color='b')
plt.scatter(x,y_pred_1,color='r')
```

```python
# now applying the plynomial regression to the case 
```

```python
x_new = np.c_[x, x**2]
clf2 = LinearRegression()
clf2.fit(x_new, y)
y_pred_2 = clf2.predict(x_new)
clf2.coef_, clf2.intercept_
```

```python
plt.scatter(x,y,color='b')
plt.scatter(x,y_pred_1,color='r')
plt.scatter(x,y_pred_2,color='y')
# now the yellow line represents the peffect fitting of the data
# regularization can still be apllied 
```


# Polynomial regression through Sklearn

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly=PolynomialFeatures(degree=2,include_bias=False)
x_poly1=poly.fit_transform(x)
```

```python
x_poly1.shape
```

```python
linear_reg=LinearRegression()
```

```python
linear_reg.fit(x_poly1,y)
```

```python
linear_reg.intercept_,linear_reg.coef_
```

```python
y_pred3=linear_reg.predict(x_poly1)
```
